Question1: 
----------
Write a Python program that takes a student's points as input and determines their corresponding letter grade based on the following grading scale:

A: 85 points and above

B: 70 points and above (but less than 85)

C: 60 points and above (but less than 70)

D: 45 points and above (but less than 60)

F: Less than 45 points


Expected Output:
--------------
Enter point: 78


The grade is B

```python
#Answer1: Write your code for Question1 here:
def LetterGrade(num):
    
    if num<45:
        grade="F"
    elif num<60:
        grade="D"
    elif num<70:
        grade="C"
    elif num<85:
        grade="B"
    else:
        grade="A"
    return grade

x= int(input("Enter point: "))
grade=LetterGrade(x)
print("The grade is ", grade)
```

Write a Python program to create a monthly duty schedule for a staff list. The staff members are Ahmet, Ali, Aye, Gkhan, Zeynep, and Fatih. The schedule should assign each staff member to a specific day of April. 
The assignment is cyclic, meaning it repeats every 6 days. 
For each day in April (from 1st to 30th), print the day, the month ("April"), and the staff member responsible for that day.

Expected output: 
----------


```python
#Answer2: Write your code for Question2 here:
staff=["Ahmet", "Ali", "Aye", "Gkhan", "Zeynep", "Fatih"]
days=30

def staff_day(day):
    index = (day - 1) % len(staff)
    return staff[index]

for day in range(1, days+ 1):
    staff_member = staff_day(day)
    print(f"{staff_member}  is the security guard on April ",day)

```
